# 파이썬으로 데이터 정제하기

```python
import pandas
# 판다스 라이브러리를 불러온다
```

```python
dataframe = pandas.read_excel('실습_정제_파이썬.xlsx')
# 판다스 라이브러리의 read_excel 함수를 사용해서 '실습_정제_파이썬.xlsx' 파일을 가져온다
# read_excel 함수가 반환하는 DataFrame 객체를 변수 dataframe에 할당
```

```python
dataframe.head()
```

```python
dataframe1 = dataframe.loc[::2]  # dataframe의 홀수행만 선택해 dataframe1에 할당
dataframe1.head(3)
```

```python
dataframe2 = dataframe.loc[1::2]  # dataframe의 짝수행만 선택해 dataframe1에 할당
dataframe2.head(3)
```

```python
dataframe1 = dataframe1.reset_index()  
# dataframe1의 행 이름을 0부터 1씩 커지는 정수로 초기화
dataframe1 = dataframe1.rename({'생년월일/주소':'생년월일'}, axis=1)
# dataframe1의 열 이름을 변경(axis=1)
# 특히 '생년월일/주소' 열 이름을 '생년월일'로 변경
# 변경한 결과를 다시 dataframe1에 할당

dataframe2 = dataframe2.reset_index()
dataframe2 = dataframe2.rename({'생년월일/주소':'주소'}, axis=1)
```

```python
dataframe3 = pandas.concat([dataframe1, dataframe2], axis=1)
# concat 함수를 써서 dataframe1과 dataframe2를 합침
# 합칠 데이터프레임 목록은 리스트(list)로 전달
# 데이터를 아래로 쌓지 않고, 오른쪽 열 방향으로 합침(axis=1)
dataframe3.head(3)
```

```python
dataframe3 = dataframe3.dropna(axis=1, how='all')
# dropna는 결측값(NaN)이 있는 데이터프레임의 행이나 열을 삭제하는 메소드
# axis=1 로 쓰면 열을 삭제
# how='all' 로 쓰면 모든 값이 결측값인 행/열만 삭제
dataframe3.head(3)
```

```python
dataframe3['시도명'] = dataframe3['시도명'].fillna(method='ffill')
# 열을 하나만 선택하고 fillna 메소드를 써서 결측값을 채움
# method='ffill'로 쓰면, 결측값에 바로 위에 있던 값을 입력
dataframe3.head(3)
```

```python
dataframe3.to_excel('정제완료.xlsx')  # dataframe3을 엑셀파일로 저장
```
