```python
import pandas
```

```python
donation = pandas.read_excel('실습_분석_국회의원고액후원금.xlsx')
# 판다스 라이브러리의 read_excel 함수를 사용해서 '실습_분석_국회의원고액후원금.xlsx' 파일을 가져온다
# read_excel 함수가 반환하는 DataFrame 객체를 변수 donation에 할당
```

```python
donation.head()
```

```python
donation[donation['후원자_성명'] == '이영애']
# '후원자_성명'이 '이영애'와 일치하는 행만 선택
```

```python
donation[(donation['후원자_성명'] == '이영애') & (donation['생년'] == 1971)]
# '후원자_성명'이 '이영애'와 일치하고, '생년'은 1971인 행만 선택
```

```python
donation['기부금액_정제'].sum()  # 기부금액_정제 열의 합계를 계산
```

```python
donation['기부금액_정제'].count()  # 기부금액_정제 열에서 비어있지 않은 값의 수를 계산
```

```python
donation['후원회id'].nunique()  # 후원회id 열에서 고유한 값의 수를 계산
```

```python
donation.groupby('소속정당_정제').sum()
# groupby 명령어를 써서 소속정당별로 값을 요약
# 뒤에 sum 을 붙여서 숫자형 변수들의 합계를 계산
# sum 자리에 count, mean, nunique 등을 쓸 수 있음
```

```python
donation.groupby('후원회id').sum().sort_values(by='기부금액_정제', ascending=False)
# groupby 명령어를 써서 후원회별로 합계를 계산하고
# 기부금액_정제 열을 기준으로 내림차순 정렬
```

```python
donation.groupby(['기부월']).sum()
```
